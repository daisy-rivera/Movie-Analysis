## Using Glob to Load & Concatenate Files

We are going to load all of our extracted API files for years 2000-2022 and combine them into one.

### Import Libraries

```python
import pandas as pd
#helps find file paths that match a specific criterion
import glob
```

```python
#find and create a list of the files sorted
q = "Data/final_movies*.csv.gz" ## change to file name and add * to represent any nuumber of other characters in the search query
chunked_files = sorted(glob.glob(q))
# Showing the first 5 
chunked_files[:5]
```

### Combining Many Files

```python
## Loading and Concatenating the list of files with 1 line
df_combined = pd.concat([pd.read_csv(file) for file in chunked_files])
#display dataframe
df_combined
```

When combining our files, we notice that an id# 0  was created at the start of each of the files merged. We want to remove all those rows and reset the index.

```python
#remove the all ids that are 0
df_combined = df_combined.loc[df_combined['imdb_id'] != '0']
df_combined.head(3)
```

```python
#reset the index
#drop=True means we are dropping the old index
df_combined = df_combined.reset_index(drop=True)
#display dataframe
df_combined.head(3)
```

```python
## Saving the final combined dataframe
final_fname ='Data/FINAL_COMBINED_MOVIES.csv.gz'
df_combined.to_csv(final_fname, compression='gzip', index=False)
```

```python
#uploading final combined file to verify
final_movies_df = pd.read_csv(final_fname)
final_movies_df.head(3)
```
